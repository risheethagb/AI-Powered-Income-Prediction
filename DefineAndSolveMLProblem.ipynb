```python
#Handle missing values by replacing them with the mean
df['hours-per-week'].fillna(df['hours-per-week'].mean(), inplace=True)
df['age'].fillna(df['age'].mean(), inplace=True)

#Fix the workclass to simplify to only two total groups
columns_not_self_employed = ~(df['workclass'] == 'Self-emp-not-inc') & ~(df['workclass'] == 'Self-emp-inc')
columns_not_null = ~(df['workclass'].isnull())  
condition = columns_not_self_employed & columns_not_null
df['workclass'] = np.where(condition, 'Not-self-emp', df['workclass'])
condition = (df['workclass'] == 'Self-emp-not-inc') | (df['workclass'] == 'Self-emp-inc')
df['workclass'] = np.where(condition, 'Self-emp', df['workclass'])

# Transform some of the categorical values that I will use to binary 
df_education = pd.get_dummies(df['education'], prefix='education_')
df = df.join(df_education)
df.drop(columns = 'education', inplace=True)

df_occupation = pd.get_dummies(df['occupation'], prefix='occupation_')
df = df.join(df_occupation)
df.drop(columns = 'occupation', inplace=True)

# Drop columns that do not add value (the capital-loss, gain)
df = df.drop(columns=['capital-gain', 'capital-loss', 'fnlwgt'])

# We can also drop native-country since there are alot unique countries
df = df.drop(columns=['native-country'])
```

```python
# I will now one-hot encode all categorical varible to binary so that I can fit the dataframe
df = pd.get_dummies(df, columns=[
    'workclass', 'marital-status', 'relationship', 'race', 'sex_selfID'
])
```

```python
# Now I will make sure all features at fittable 
df.dtypes
```

```python
# Now I will create the training and testing data sets
y = df['income_binary'] 
X = df.drop(columns = 'income_binary', axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1234)
```

```python
# Here I will create the parameter grid and will attempt to perform model selection 
param_grid = {
    'n_estimators': [20, 50, 100],
    'max_depth': [2**n for n in range(2,6)],
    'min_samples_leaf': [2**n for n in range(0,5)]
}

# Create the base RF model 
base_model = RandomForestClassifier(criterion='entropy')

# Perfrom grid search with the created dictionary above
grid = GridSearchCV(estimator=base_model, param_grid=param_grid, cv=5)
grid_search = grid.fit(X_train, y_train)

# Compute the best hyperparameters config and score
print('Optimal hyperparameters: {0}'.format(grid_search.best_params_))
print('Accuracy score: {0}'.format(grid_search.best_score_))

# Choose the best model
best_model = grid_search.best_estimator_
best_model.fit(X_train, y_train)

# Make predictions on the test data 
rf_prediction = best_model.predict_proba(X_test)
rf_predictions_list = list(rf_prediction[:, -1])






```

Based on the above, we see that after model selection, the best parameters for the Random Forest is 'max_depth': 32, 'min_samples_leaf': 25, 'n_estimators': 100. Which gave us a pretty high accuracy score of 0.837176254870502. 

```python
# I will now evaluate my performance using AUC and ROC for the graph to visually show that the 100 estimators hyperparameter is pretty high. 

# Use roc_curve to record fpr and tpr for rf100
fpr_100, tpr_100, thresholds_100 = roc_curve(y_test, rf_predictions_list)

```

```python
# This will now plot the roc for all three models
fig = plt.figure()
ax = fig.add_subplot(111)

sns.lineplot(x=fpr_100, y=tpr_100, marker = 'o')


plt.title("Receiver operating characteristic (ROC) curve")
plt.xlabel("False positive rate")
plt.ylabel("True positive rate")
plt.legend(['RF with 100 estimators'])
plt.show()
```

I see that the model that was fit using a larger number of estimators performed better. I will now quantify this with AUC to calculate the area under the ROC curve

```python

auc_100 = auc(fpr_100, tpr_100)
print("AUC of the RF model with 100 estimators is {:.3f}".format(auc_100))
```

The AUC of 0.875 is pretty high and shows us that the predictions were pretty accurate. This shows us that a Random Forest with the hyperparameters computed in the above steps were nicely optimized.
